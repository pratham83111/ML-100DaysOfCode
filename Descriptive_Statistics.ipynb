```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cleaned_employee_data.csv")
```

```python
# Random dataset with 20,000 rows
np.random.seed(42)
df = pd.DataFrame({
    'Age': np.random.randint(18, 60, 20000),
    'Salary': np.random.randint(20000, 150000, 20000),
    'Experience': np.random.randint(0, 20, 20000),
    'Gender': np.random.choice(['Male', 'Female'], 20000)
})

df.head()
```

```python
df.describe()
```

```python
sns.boxplot(x=df['Salary'])
plt.title("Salary Distribution (with Outliers)")
plt.show()
```

```python
df.describe()
```

```python
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title("Gender Distribution")
plt.show()
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Between Features")
plt.show()
```
