```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Notebook display settings
plt.style.use('ggplot')
```

```python
# Agar data.csv file hai to:
df = pd.read_csv('employee_data.csv')

# Pehle 5 rows dekho
df.head()
```

```python
df.shape          # total rows and columns
df.info()         # column data types and null info
df.describe()     # numerical columns ke summary stats
```

```python
df.isnull().sum()

# Example: Fill missing Age with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

# Example: Fill missing Department with mode (most common value)
df['Department'].fillna(df['Department'].mode()[0], inplace=True)
```

```python

cat_cols = ['Gender', 'Department', 'City']


df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

df.head()
```

```python
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title("Age Distribution of Employees")
plt.show()
```

```python
sns.countplot(x='Gender', data=df, palette='Set2')
plt.title("Gender Distribution")
plt.show()
```
