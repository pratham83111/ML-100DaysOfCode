```python
# Coin Toss Probability

favorable_outcomes = 1  # Head
total_outcomes = 2      # Head, Tail

probability_of_head = favorable_outcomes / total_outcomes
print("Probability of getting Head:", probability_of_head)
```

```python
# Dice Roll Probability

favorable_outcomes = 3  # Even numbers: 2, 4, 6
total_outcomes = 6

probability_even = favorable_outcomes / total_outcomes
print("Probability of getting an even number:", probability_even)
```

```python
import random

trials = 10000
success = 0

for _ in range(trials):
    outcome = random.choice(["H", "T"])
    if outcome == "H":
        success += 1

print("Experimental Probability of getting Head:", success/trials)
```
