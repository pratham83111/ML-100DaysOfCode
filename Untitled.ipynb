```python

import pandas as pd
import numpy as np


df = pd.read_csv("employee_data.csv")

df.head()
```

```python

print("Shape of dataset:", df.shape)
df.info()
df.describe()
```

```python
# Check missing values count
df.isnull().sum()

# Example: Fill missing Age with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Salary'] = df['Salary'].fillna(df['Salary'].median())

# Example: Fill missing Department with mode (most common value)
df['Department'].fillna(df['Department'].mode()[0], inplace=True)
```

```python

cat_cols = ['Gender', 'Department', 'City']


df = pd.get_dummies(df, columns=cat_cols, drop_first=True)

df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()


num_cols = ['Age', 'Experience', 'Salary', 'PerformanceScore']
df[num_cols] = scaler.fit_transform(df[num_cols])

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.boxplot(df['Salary'])
plt.title("Salary Outlier Visualization")


Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['Salary'] >= Q1 - 1.5*IQR) & (df['Salary'] <= Q3 + 1.5*IQR)]

print("New shape after removing outliers:", df.shape)
```

```python
df.to_csv("cleaned_employee_data.csv", index=False)
print("✅ Cleaned dataset saved successfully!")
```
